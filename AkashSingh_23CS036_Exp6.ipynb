```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing import sequence

print("Libraries imported successfully!")
```

```python
max_features = 10000
maxlen = 500

print('Loading data...')
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_features)

# Since reviews have different lengths, we pad them to ensure they are all 500 words long
x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
x_test = sequence.pad_sequences(x_test, maxlen=maxlen)

print(f'Train shape: {x_train.shape}')
print(f'Test shape: {x_test.shape}')
```

```python
model = Sequential()

# Layer 1: Embedding (Turns word integers into dense vectors)
model.add(Embedding(max_features, 32))

# Layer 2: SimpleRNN (The recurrent part that remembers sequence order)
model.add(SimpleRNN(32))

# Layer 3: Dense Output (Sigmoid gives a probability between 0 and 1)
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['acc'])

model.summary()
```

```python
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=128,
                    validation_split=0.2)
```

```python
results = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {results[1]*100:.2f}%")
```

```python
def predict_sentiment(text):
    word_to_index = imdb.get_word_index()
    # Tokenize and offset by 3 (IMDB convention)
    words = text.lower().split()
    review = [word_to_index.get(w, 2) + 3 for w in words]
    review = [i if i < max_features else 2 for i in review]

    padded_review = sequence.pad_sequences([review], maxlen=maxlen)
    prediction = model.predict(padded_review)[0][0]

    sentiment = "Positive 😊" if prediction > 0.5 else "Negative 😞"
    return f"Score: {prediction:.4f} | Sentiment: {sentiment}"

# Try it out!
my_review = "this movie was a masterpiece the acting was incredible"
print(predict_sentiment(my_review))
```
